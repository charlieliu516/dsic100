## **Data Description**

There are 196 observations in the players dataset and 1535 observations in the sessions dataset.

### players dataset

+ experience (chr) — describes the player’s experience level, with possible values: Pro, Veteran, Amateur, Regular, and Beginner.

+ subscribe (lgl) — indicates whether the player has subscribed to receive game updates; values are TRUE or FALSE.

+ played_hours (dbl) — total number of gameplay hours recorded for each player.

+ name (chr) — the player’s name (non-unique identifier, for reference).

+ gender (chr) — describes the player’s gender identity (e.g., Male, Female, Non-binary, Agender, Other, Two-Spirited, Prefer not to say).

+ Age (dbl) — player’s age in years.

### sessions dataset

+ hashedEmail (chr) — hashed email address used to link session data to the corresponding player in the players dataset.

+ start_time (chr) — the recorded start time of a game session (in dd/mm/yyyy hh:mm format).

+ end_time (chr) — the recorded end time of the same session.

+ original_start_time (dbl) — the timestamp representation of start_time (in milliseconds since epoch).

+ original_end_time (dbl) — the timestamp representation of end_time (in milliseconds since epoch).

## **potential issues**

1.Categorical variables such as experience and gender are non-numeric.
They will need to be encoded before being used as predictors in a model.  
2.played_hours and Age may contain missing values (NA), which need to be handled before statistical analysis.  
3.sessions dataset uses time columns (start_time, end_time) stored as characters instead of datetime objects.
