## **Importing data**

```r
library(tidyverse)
url = 'https://drive.google.com/uc?export=download&id=1Nkmsri93Yq5Ind2ztUNSQlzEle0Xzde0'
players <- read_csv(url)
head(players)
```

```r
url2 = 'https://drive.google.com/uc?export=download&id=1q0FAgI3CWPZTFz8lLgFK6WOQNOfsYrGS'
sessions <- read_csv(url2)
head(sessions)
```

```r
distinct(players, experience)
distinct(players, subscribe)
distinct(players, gender)
players |>
  summarize(
    max_played = max(played_hours, na.rm = TRUE),
    min_played = min(played_hours, na.rm = TRUE),
    max_age    = max(Age, na.rm = TRUE),
    min_age    = min(Age, na.rm = TRUE))
```

## **Data Description**

There are 196 observations in the players dataset and 1535 observations in the sessions dataset.

### players dataset

+ experience (chr) — describes the player’s experience level, with possible values: Pro, Veteran, Amateur, Regular, and Beginner.

+ subscribe (lgl) — indicates whether the player has subscribed to receive game updates; values are TRUE or FALSE.

+ played_hours (dbl) — total number of gameplay hours recorded for each player.

+ name (chr) — the player’s name (non-unique identifier, for reference).

+ gender (chr) — describes the player’s gender identity (e.g., Male, Female, Non-binary, Agender, Other, Two-Spirited, Prefer not to say).

+ Age (dbl) — player’s age in years.

### sessions dataset

+ hashedEmail (chr) — hashed email address used to link session data to the corresponding player in the players dataset.

+ start_time (chr) — the recorded start time of a game session (in dd/mm/yyyy hh:mm format).

+ end_time (chr) — the recorded end time of the same session.

+ original_start_time (dbl) — the timestamp representation of start_time (in milliseconds since epoch).

+ original_end_time (dbl) — the timestamp representation of end_time (in milliseconds since epoch).

## **potential issues**

1.Categorical variables such as experience and gender are non-numeric.
They will need to be encoded before being used as predictors in a model.  
2.played_hours and Age may contain missing values (NA), which need to be handled before statistical analysis.  
3.sessions dataset uses time columns (start_time, end_time) stored as characters instead of datetime objects.

## **Data visualization**

```r
ggplot(players, aes(x = subscribe)) +
  geom_bar() +
  facet_grid(cols = vars(experience))
```

```r
ggplot(players, aes(x = subscribe, y = played_hours)) +
  geom_boxplot()
```

```r
players |> filter(played_hours <= 10, Age <= 50) |>
  ggplot(aes(x = played_hours, y = Age, color = subscribe)) +
  geom_point()
```

```r
ggplot(players, aes(x = subscribe)) +
  geom_bar() +
  facet_grid(cols = vars(experience))
```

```r
ggplot(players, aes(x = played_hours, y = Age, color = subscribe)) +
  geom_point(alpha = 0.6)
```

From the visualizations, we can observe that players with higher experience levels (such as Pro and Veteran) are more likely to subscribe, and they also tend to have longer playtime compared to Beginner players.  
Most players have relatively short playtime (under 10 hours), while a few show extremely high values, indicating a right-skewed distribution.
Age does not show a clear relationship with either subscription or playtime.  
Additionally, gender seems to have some influence, as male and non-binary players exhibit higher maximum playtime.
